```python
from langchain.text_splitter import CharacterTextSplitter
splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size=200,
    chunk_overlap=0
)

chunks = splitter.split_text(text)
print(chunks)


```

```python
for chunk in chunks:
  print(len(chunk))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
r_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n","\n"," "],
    chunk_size=200,
    chunk_overlap=0
)
chunks = r_splitter.split_text(text)
for chunk in chunks:
  print(len(chunk))
```

```python
from langchain import OpenAI
llm = OpenAI(temperature=0.8)
```

```python
loader = UnstructuredURLLoader(
    urls = [
        "https://www.moneycontrol.com/news/business/banks/hdfc-bank-re-appoints-sanmoy-chakrabarti-as-chief-risk-officer-11259771.html",
        "https://www.moneycontrol.com/news/business/markets/market-corrects-post-rbi-ups-inflation-forecast-icrr-bet-on-these-top-10-rate-sensitive-stocks-ideas-11142611.html"
    ]
)
```

```python
data = loader.load()
print(data[0])
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap=200
)
docs = text_splitter.split_documents(data)
len(docs)
```

```python
!pip install tiktoken
```

```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
embeddings = OpenAIEmbeddings()
vectorindex_openai = FAISS.from_documents(docs,embeddings)
```

```python
import pickle
with open("vector_index.pkl",'wb') as a:
  pickle.dump(vectorindex_openai, a)
```

```python
with open("vector_index.pkl",'rb') as a:
  vector_index = pickle.load(a)
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
chain = RetrievalQAWithSourcesChain.from_llm(llm =llm,retriever = vector_index.as_retriever())
```

```python
query1 = "What does Osho Krishan feels on Nifty50"
# langchain.debug = True
chain({'question':query1},return_only_outputs=True)
```
